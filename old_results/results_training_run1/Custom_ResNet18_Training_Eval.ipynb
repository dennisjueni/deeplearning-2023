```python
import numpy as np

choices_epochs = [40,41,42]

best_seed_val_accs = []
best_seed_train_accs = []

for i in range(0,20):
    for j in range(i+1,20):
        max_val_acc = 0
        max_train_acc = 0
        for epoch in choices_epochs:
            tracked_params = np.load(f"./models/tracked_params{i}_{j}_{epoch}.npy", allow_pickle=True).item()
            max_val_acc = max(float(tracked_params["val_acc_top1"][-1]),max_val_acc)
            max_train_acc = max(float(tracked_params["train_acc_top1"][-1]),max_train_acc)
            
        best_seed_val_accs.append(max_val_acc)
        best_seed_train_accs.append(max_train_acc)
        

# plot histogram of train and val accs
import matplotlib.pyplot as plt
plt.hist(best_seed_val_accs, bins=100)
# set title and labels
plt.title('Histogram of best val accs')
plt.xlabel('val acc')
plt.ylabel('count')
plt.savefig('best_val_accs.png')
plt.clf()
# repeat for train accs
plt.hist(best_seed_train_accs, bins=100)
# set title and labels
plt.title('Histogram of best train accs')
plt.xlabel('train acc')
plt.ylabel('count')
plt.savefig('best_train_accs.png')
plt.clf()

# compute the avg and std of the best accs
print("avg val acc", np.mean(best_seed_val_accs))
print("std val acc", np.std(best_seed_val_accs))
print("avg train acc", np.mean(best_seed_train_accs))
print("std train acc", np.std(best_seed_train_accs))
```
